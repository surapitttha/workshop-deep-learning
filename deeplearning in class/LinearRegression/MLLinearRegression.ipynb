```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [1.0,1.8,3.0,4.1,5.2,6.0]
y_data = [1,  1.3,2.2,2.5,2.8,3.6]

x = np.array(x_data) #Numpy Array
y = np.array(y_data)
```

```python
x
```

```python
y
```

```python
#data Visualization

plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
x= x.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x ,y)
```

```python
#predict
x_input = 2.5
model.predict([[x_input]])
```

```python
x_input = [2.0,2.5,3,7.0,10]
x_input = np.array(x_input)
x_input = x_input.reshape(-1,1)
y_pred = model.predict(x_input)
y_pred
```

```python
prd = model.predict(x)
plt.grid()
plt.scatter(x_data,y_data,color='b')
plt.plot(x_data,prd,linewidth='1')
plt.scatter(x_data,prd,color='r',marker='s')
plt.plot()
```
