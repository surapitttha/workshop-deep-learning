```python
import joblib
import numpy as np

model=joblib.load('myfirstLinearRegression model')
```

```python
model.coef_ ,model.intercept_
```

```python
x_input= [4,5,6,7,8,9,10]
x_input = np.array(x_input).reshape(-1,1)

ypred = model.predict(x_input)
ypred
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.scatter(x_input,ypred,marker='s')
plt.plot(x_input,ypred,color='red')
plt.show()
```
