```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data set
x = [27,35,39,21,22,45]
x = np.array(x).reshape(-1,1)

y_name = ['lemon','orenge','orenge','lemon','lemon','orenge']
y = [0,1,1,0,0,1]
```

```python
#data visualization

plt.xlabel('fruit')
plt.ylabel('size')
plt.scatter(y_name,x)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
x_input = [[25.3]]
ypred = model.predict(x_input)
ypred
```

```python
fruit_classes = ['lemon','Orange']
```

```python
prd = fruit_classes[ypred[0]]
print('Result:{}'.format(prd))
```

```python
x_input = [31,25,21.2,28.6,38]
x_input = np.array(x_input).reshape(-1,1)

ypred = model.predict(x_input)
ypred
```

```python
print('Result: ')
for i in ypred :
    print(fruit_classes[i])
```
