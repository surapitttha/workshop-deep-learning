```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import plot_tree
from mlxtend.plotting import plot_decision_regions
from sklearn.tree import DecisionTreeClassifier
```

```python
#Datasets
x = [27,35,39,21,22,45,102,110,120]
x = np.array(x).reshape(-1,1)

y_name = ['lemon','orange','orange','lemon','lemon','orange','pamelo','pamelo','pamelo']
y = [0,1,1,0,0,1,2,2,2]
```

```python
#Data Visualization
plt.xlabel('fruit')
plt.ylabel('size')

plt.scatter(y_name,x)
plt.show()
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
fruit_classes = ['lemon','Orange','Pamelo']
```

```python
x_input= np.array([102]).reshape(-1,1)
ypred = model.predict(x_input)

prd = fruit_classes[ypred[0]]
print('Result : {}'.format(prd))
```

```python
x_input = np.array([28,32,39,101]).reshape(-1,1)

y_pred = model.predict(x_input)

y_pred
```

```python
for i,k in enumerate(y_pred):
    print(f'{x_input[i]} = {fruit_classes[k]}')
```

```python
#Visualize model
data_freature_name = ['size']
plt.subplots(nrows=1,ncols=1,figsize=(4,3),dpi=96)
plot_tree(model,
         filled=True,rounded=True,
        feature_names=data_freature_name,
        class_names=fruit_classes,)
plt.show()
```

```python
#plot needs Numpy
x= np.array(x)
y = np.array(y)
```

```python
#plot decision regions
plot_decision_regions(x,y,clf=model)
plt.show()
```
